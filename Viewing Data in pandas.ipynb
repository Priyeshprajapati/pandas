```python
# importing pandas as pd
import pandas as pd
import numpy as np
# Creating the Series
sr = pd.Series(['New York', 'Chicago', 'Toronto', 'Lisbon', 'Rio'])
  
# Create the Index
index_ = ['City 1', 'City 2', 'City 3', 'City 4', 'City 5'] 
  
# set the index
sr.index = index_
  
# Print the series
print(sr,'\n\n\n')

# return numpy array representation
result =np.array(sr)
  
# Print the result
print(result)
```

```python
# importing pandas as pd
import pandas as pd
  
# Creating the Series
sr = pd.Series([11, 21, 8, 18, 65, 18, 32, 10, 5, 32, None])
  
# Create the Index
# apply yearly frequency
index_ = pd.date_range('2010-12-29 08:45', periods = 11, freq ='y')
  
# set the index
sr.index = index_
  
# Print the series
print(sr)
# return numpy array representation
result = np.array(sr)
  
# Print the result
print(result)
```
