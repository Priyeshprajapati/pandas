```python
weather_d={
    'days':['1/1/2017','1/6/2017','1/4/2017'],
    'temperature':[32,35,28],
    'windspeed':[6,2,8],
    'evnet':['rain','sunny','snow']
}
df=pd.DataFrame(weather_d)
df
```

```python
weather_d=[
    ('1/1/2017',32,6,'rain'),
    ('1/2/2017',22,7,'sunny'),
    ('1/3/2017',34,9,'snow')
]
df=pd.DataFrame(weather_d,columns=['day','temp','windspeed','event'])
df
```
